```python
nltk.download("stopwords")
```

```python
# Below are the stopwrods in english :
print(stopwords.words("english"))
```

```python
# lets find how many stopwords are present
len(stopwords.words("english"))
```

* we see that the word "Comcast"is used so much, which is not a problem case but a name of the company,so we dont need it and other words like ("," and "-" and few other lets also remove them 
* These are called the punctuation in the english we will remove all that 

```python
## puncuation markas
import string # library for removing punctuations
punc=string.punctuation
print(punc)
```

```python
punc=list(punc)
punc
```

```python
# lets make new list of my stopwords :
stop_words=stopwords.words("english")
my_word=['comcast']
my_stopwords=stop_words+punc+my_word
```

```python
print(my_stopwords)
```

* we can see that all words are included

```python
#### lets filter all the words by my_Stopwords
filterd_comp=[word for word in words if not word in my_stopwords]
```

```python
len(filterd_comp)
```

* we have 6872 filterdwords

### stemming 
- coverting the inflextions into the root words by removing the affxes of the words like(-ed,-ing,-es etc..)
- stemming just done prefix or suffix of the word and covert into a root word.
- Stemming always does not give the actual word .

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
stem_comps=[]
for word in filterd_comp:
    stem_comps.append(ps.stem(word))
```

```python
print(stem_comps)
```

* Few words in the list may not make any scence or have any meaning , it is due to the portstemmer

```python
from nltk.probability import FreqDist
fd = FreqDist(stem_comps)
print(fd)
```

```python
fd.plot(50, cumulative = False)
plt.title("Freqency of the top 50 words")
```
